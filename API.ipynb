Rick and Morty API

```python
import pandas as pd
import requests
```

```python
url1 = "https://rickandmortyapi.com/api"
endpoint = "/character"
requests.get(url1+endpoint)
```

```python
def  page_request(url1,endpoint, page):
    response = requests.get(url1+endpoint+f'?page={page}')
    return response.json()
```

```python
def get_pages(response):
    return response['info']['pages']
```

```python
#get information
def parse_json(response):
    character_list = []
    for item in response['results']:
        character ={
        'name':item['name'],
        'species': item['species'],
        'Number_of_Species': len(item['episode'])
        }
        character_list.append(character)
    return character_list

```

```python
# We want to know how many pages exist
data = page_request(url1,endpoint,1)

# Get the total pages
total_pages = get_pages(data)

# Loop through all pages and get the characters
Characters_List_Main = []

for x in range(1,total_pages + 1):
    response = page_request(url1,endpoint,x)
    Characters_List_Main.extend(parse_json(response))
```

```python
print(len(Characters_List_Main))
Characters_List_Main[:5]
```

```python
pd.DataFrame(Characters_List_Main).head()
```
