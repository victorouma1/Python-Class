Rick and Morty API

```python
import pandas as pd
import requests
```

```python
url1 = "https://rickandmortyapi.com/api"
endpoint = "/character"
requests.get(url1+endpoint)
```

```python
def  page_request(url1,endpoint, page):
    response = requests.get(url1+endpoint+f'?page={page}')
    return response.json()
```

```python
def get_pages(response):
    return response['info']['pages']
```

```python
#get information
def parse_json(response):
    character_list = []
    for item in response['results']:
        character ={
        'name':item['name'],
        'species': item['species'],
        'Number_of_Species': len(item['episode'])
        }
        character_list.append(character)
    return character_list

```

```python
# We want to know how many pages exist
data = page_request(url1,endpoint,1)

# Get the total pages
total_pages = get_pages(data)

# Loop through all pages and get the characters
Characters_List_Main = []

for x in range(1,total_pages + 1):
    response = page_request(url1,endpoint,x)
    Characters_List_Main.extend(parse_json(response))
```

```python
print(len(Characters_List_Main))
Characters_List_Main[:5]
```

```python
pd.DataFrame(Characters_List_Main).head()
```

```python
import requests
import pandas
import time
import csv
```

```python
Movie_Input = input("Enter a movie: ") 
Movie_Search =  Movie_Input.replace(" ", "+")
if Movie_Input == "":
    print("Empty input, Enter a movie")
```

```python
API_Key = "60ec1ab7"
Movie = []

Movie_URL = f"http://www.omdbapi.com/?t={Movie_Search}&apikey={API_Key}" 
Movie_response = requests.get(Movie_URL)
Movie_response
```

```python
if Movie_response == "<Response [400]>":
    print("Invalid or missing API")
if Movie_response == "<Response [204]>":
    print("Movie not found")
```

```python
Movie = Movie_response.json()
Movie
```

```python
Criteria = ['Title', 'Year', 'Genre', 'Director', 'Plot', 'imdbRating']
Details = []
for movies in Movie:
    for i in Criteria:
        if movies == i:
            Details.append(Movie[movies])

for j in range(len(Details)):
    print(f"{Criteria[j]}: {Details[j]}")

with open('Movies.csv','w',newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Criteria', 'Details'])
    for x in range(len(Details)):
        writer.writerow([Criteria[x], Details[x]])



```
