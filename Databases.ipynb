Create a new Database and connect to it

```python
conn = sqlite3.connect("school1.db")
conn.execute("PRAGMA foreign_keys = ON") # Enforces Foreign Key Constraints
```

```python
cursor = conn.cursor()
```

Create the students Table

```python
cursor.execute("""
            CREATE TABLE IF NOT EXISTS Students(
               student_id INTEGER PRIMARY KEY AUTOINCREMENT,
               first_name TEXT NOT NULL,
               last_name TEXT NOT NULL,
               age INTEGER NOT NULL,
               grade TEXT NOT NULL,
               email TEXT NOT NULL UNIQUE
               )                        
               """)
conn.commit() # save the changes to the database
```

Add a new column

```python
cursor.execute("ALTER TABLE students ADD COLUMN phone_number INTEGER")
conn.commit()
```

Deleting a column

```python
cursor.execute("ALTER TABLE students DROP COLUMN phone_number")
conn.commit()
```

Using Foreign Keys

```python
import sqlite3
```

```python
cursor.execute("""  
               CREATE TABLE IF NOT EXISTS enrollments(
                    enrollment_id INTEGER PRIMARY KEY AUTOINCREMENT,
                    student_id INTEGER NOT NULL,
                    course_id INTEGER NOT NULL,
                    FOREIGN KEY (student_id) REFERENCES Students(student_id) ON DELETE CASCADE,
                    FOREIGN KEY (course_id) REFERENCES Courses(course_id) ON DELETE CASCADE
               ) 
               """)
conn.commit()
```

```python
print(cursor.execute("SELECT sql FROM sqlite_master WHERE name='Students'").fetchone())
print(cursor.execute("SELECT sql FROM sqlite_master WHERE name='Courses'").fetchone())
```

```python
enrollments_data = [
    (2,1),
    (3,2),
    (2,3),
    (4,1),
    (5,4),
]

cursor.executemany("INSERT OR IGNORE INTO enrollments (student_id, course_id) VALUES(?, ?)", enrollments_data)
conn.commit()
```

```python
cursor.execute("""
               SELECT
               Students.first_name, Students.last_name AS student_name,
               Courses.course_name AS course_name
               FROM enrollments
               JOIN Students
                    ON enrollments.student_id = Students.student_id
               JOIN Courses
                    ON enrollments.course_id = Courses.course_id
               """)
enrollments = cursor.fetchall()

for enrollment in enrollments:
    print(enrollment)
```
