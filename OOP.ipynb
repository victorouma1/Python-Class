```python
# Using Multiple Parameters
class Car:
    def __init__(self, make, model, year, colour):
        self.make = make
        self.model = model
        self.year = year
        self.colour = colour
    
    def start_engine(self):
        print(f"The {self.colour} {self.year} {self.make} {self.model} is now running")

C1 = Car("Audi","R8", 2020, "Red")
C2 = Car("Ford", "Ranger", 2005, "Black")
C1.start_engine()
C2.start_engine()
```

Class Attributes vs Instance Attributes

```python
class Circle:
    pi = 3.143 # class attribute
    def __init__ (self,radius):
        self.radius = radius # instance attribute
    def Area(self):
        return self.pi * self.radius ** 2

Circle1 = Circle(2)
Circle1.Area()
```

```python
Circle.pi = 3.14
```

Mutator Methods

```python
class Product:
    def __init__ (self, price):
        self.price = price
    
    def set_discount(self,amount):
        self.price -= amount
        print("Discount applied")
        return self.price

Item = Product(100)
print(f"Original Price: {Item.price}")
print(f"Discounted Price: {Item.set_discount(50)}")
```

Accessor Methods

```python
class Product:
    def __init__ (self, price):
        self.price = price
    
    def set_discount(self,amount):
        self.price -= amount
        print("Discount applied")
        return self.price
    def get_tax(self):
        return self.price * 0.20

Item = Product(100)
print(f"Original Price: {Item.price}")
print(f"Discounted Price: {Item.set_discount(10)}")

Tax = Item.get_tax()
print(f"The tax is {Tax}")
```

Helper Methods

```python
class Product:
    def __init__ (self, price):
        self.price = price
    
    def set_discount(self,amount):
        self.price -= amount
        print("Discount applied")
        return self.price
    def get_tax(self):
        return self.price * 0.20
    def calculate_total(self):
        # calling another method in the same class
        tax = self.get_tax()
        return self.price + tax


Item = Product(100)
print(f"Original Price: {Item.price}")
print(f"Discounted Price: {Item.set_discount(20)}")
Tax = Item.get_tax()
print(f"The tax is {Tax}")
print(f"Total price: {Item.calculate_total()}")
```

Class and Static Methods

```python
class MathOperations:
    pi = 3.14159

    @classmethod
    def Circle_Area(cls, radius):
        return cls.pi * radius ** 2
    @staticmethod
    def add(a,b):
        return a + b
    
Object1 = MathOperations()
Object2 = MathOperations()

#Using the Class method
Area_1 = Object1.Circle_Area(5)
print(f"The Area of the ciricle with radius 5 is: {Area_1}")

#Using the Static Method
sum_1 = Object1.add(5,10)
print(f"The sum is {sum_1}")
```
