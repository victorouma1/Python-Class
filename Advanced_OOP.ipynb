```python
class point:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __eq__(self,other):
        return (self.x == other.x and self.y == other.y)
    def __ne__(self,other):
        return (self.x != other.x and self.y != other.y)
    

p1 = point(3,4)
p2 = point(6,5)
print(p1 != p2)
```

```python
class Number:
    def __init__(self,value):
        self.value = value
    def __lt__(self,other):
        return (self.value < other.value)

n1 = Number(5)
n2 = Number(10)

print(n1 < n2)
```

```python
class Vector:
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __add__(self,other):
        return (self.x + other.x, self.y + other.y)
    def __str__(self):
        return f"Results(x = {self.x} y = {self.y})"

v1 = Vector(5,10)
v2 = Vector(50,100)
v3 = v1 + v2
print(v3)
```

```python
class Grandparent:
    def say_hello(self):
        print("Hello from Grandparent!")
class Parent(Grandparent):
    pass
class Child(Parent):
    def skill(self):
        print("Cooking")

child_1 = Child()
child_1.say_hello()
child_1.skill()
```

```python
class User:
    def __init__(self, name, email):
        self.name = name
        self.email = email
    
    def login(self):
        print(f"{self.name} has logged in")

    def display_info(self):
        print(f"Name: {self.name}")
        print(f"Email: {self.email}")
    
class Student(User):
    def __init__(self,name,email,student_id):
        super().__init__(name,email)
        self.student_id = student_id
        self.grades = []

    def add_grade(self,grade):
        self.grades.append(grade)

    def calculate_average(self):
        if len(self.grades) == 0:
            return 0
        return sum(self.grades)/len(self.grades)
    def display_info(self):
        super().display_info
        print(f"Student ID: {self.student_id}")
        print(f"Average Grade: {self.calculate_average():.2f}")

student1 = Student('Victor', 'example@example.com', 1234)
student1.login()
student1.add_grade(20)
student1.add_grade(50)
student1.display_info()

    
```

Abstraction

```python
from abc import ABC, abstractmethod

# abstract class
class Car(ABC):
    @abstractmethod
    def mileage(self):
        pass

class Tesla(Car):
    def mileage(self):
        print("The Mileage is 30 kilometers per hour")
    
class Toyota(Car):
    def mileage(self):
        print("The Mileage is 20 kilometers per hour")

class Suzuki(Car):
    def mileage(self):
        print("The Mileage is 25 kilometers per hour")

tesla_1 = Tesla()
toyota_1 = Toyota()
suzuki_1 = Suzuki()

tesla_1.mileage()
toyota_1.mileage()
suzuki_1.mileage()
```

Abstraction

```python
class Payment(ABC):
    @abstractmethod
    def make_payment(self, amount):
        pass

class MPesa_Payment(Payment):
    def make_payment(self, amount):
        print(f"Paid {amount} using MPesa")

class Card_Payment(Payment):
    def make_payment(self, amount):
        print(f"Paid {amount} using Credit Card")
    
mpesa_payment = MPesa_Payment().make_payment(1000)
card_payment = Card_Payment()

mpesa_payment
card_payment.make_payment(2000)
```

Polymorphism

```python
class Bird:
    def intro(self):
        print("There are many types of birds.")

    def flight(self):
        print("Most birds can fly but some cannot.")

class Sparrow(Bird):
    def flight(self):
        print("Sparrows can fly.")

class Ostrich(Bird):
    def flight(self):
        print("Ostriches cannot fly.")

# Polymorphic Function
def activate_bird(bird_obj):
    # This function works for ANY bird
    bird_obj.intro()
    bird_obj.flight()

# Usage
s = Sparrow()
o = Ostrich()

activate_bird(s) 
activate_bird(o) 
```
