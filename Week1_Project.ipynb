```python
import random
Games = int(input("Would you prefer a best of three or five"))
Computer_Options = ['Rock','Paper','Scissors']

Man = 0
Machine = 0
count = 0

while count < Games:
    User_Input = input("Enter Rock,Paper or Scissors").title()
    x = random.choice(Computer_Options) 
    print(f"You chose {User_Input}, the computer chooses {x}")
    #for i in range(Games):
    if User_Input == "Rock" and x == "Paper":
            Machine += 1
            count += 1
    if User_Input == "Rock" and x == "Scissors":
            Man += 1
            count += 1
    if User_Input == "Rock" and x == "Rock":
            count += 1
    if User_Input == "Paper" and x == "Rock":
            Man += 1
            count += 1
    if User_Input == "Paper" and x == "Scissors":
            Machine += 1
            count += 1
    if User_Input == "Paper" and x == "Paper":
            count += 1
    if User_Input == "Scissors" and x == "Rock":
            Machine += 1
            count += 1
    if User_Input == "Scissors" and x == "Paper":
            Man += 1
            count += 1
    if User_Input == "Scissors" and x == "Scissors":
            count += 1
    if count == Games:
            if Man > Machine:
                    print("You win")
            if Man < Machine:
                    print("You lose")
            if Man == Machine:
                    print("Draw")
    if (Man == 2 or Machine == 2) and Games == 3:
            if Man > Machine:
                    print("You win")
                    break
            if Man < Machine:
                    print("You lose")
                    break
    if (Man == 3 or Machine == 3) and Games == 5:
            if Man > Machine:
                    print("You win")
                    break
            if Man < Machine:
                    print("You lose")
                    break
```
